```python
# Função para obter coordenadas a partir do CEP
def get_coordinates_from_cep(cep):
    try:
        location = geolocator.geocode(cep)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        print(f"Erro de timeout ao buscar coordenadas para o CEP {cep}. Tentando novamente...")
        return get_coordinates_from_cep(cep)  # Tenta novamente em caso de timeout
    except Exception as e:
        print(f"Erro ao buscar coordenadas para o CEP {cep}: {e}")
        return None, None
```

```python
# Função para aplicar a geocodificação e exibir progresso
def geocode_with_progress(row, index, total_rows):
    latitude, longitude = get_coordinates_from_cep(row['postal_code'])
    if (index + 1) % 250 == 0:  # Exibe mensagem a cada 250 requisições
        print(f"Requisições concluídas: {index + 1} de {total_rows}")
         # Salva o dataframe em .csv a cada 250 requisições
        apt.to_csv('apt_SP_pointsadded.csv', index=False)
        print('Arquivo salvo como apt_SP_pointsadded.csv')
    time.sleep(2)  # Atraso de 2 segundo entre requisições para evitar sobrecarga
    return latitude, longitude
```

```python
# Aplicando a função para preencher as colunas de latitude e longitude
for index, row in apt.iterrows():
    latitude, longitude = geocode_with_progress(row, index, len(apt))
    apt.at[index, 'latitude'] = latitude
    apt.at[index, 'longitude'] = longitude
```
