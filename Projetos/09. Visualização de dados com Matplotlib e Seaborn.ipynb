## Matplotlib

Matplotlib é uma biblioteca de visualização gráfica de dados em Python amplamente utilizada para criar gráficos e visualizações de alta qualidade em diversas áreas. Permite a criação de representações gráficas em 2D, 3D, histogramas, diagrama de dispersão, gráfico de linhas, etc. Diversos tipos de plotagem são permitidas com configurações personalizáveis. É compatível com diversos formatos de arquivos de imagem.

https://matplotlib.org/

A criação de um plot envolve a escolha dos dados a serem plotados, a seleção do tipo de plot e a configuração das opções de plotagem, como tamanho da figura, rótulo de eixos, título, cores, etc.

```python
import matplotlib.pyplot as plt
%matplotlib inline        

#acima, comando para os gráficos aparecerem nesta janela do navegador.
```

```python
plt.plot([1,3,5],[2,4,6])            #O método plot define os eixos do gráfico.
plt.show()
```

```python
x = [-6,-4,-2,-1,0,1,2,4,6]
y = [36,16,4,1,0,1,4,16,36]
```

```python
plt.plot(x,y, label = "$f(x) = x^2$")       #label = legenda dos dados
plt.xlabel('x')                             #label eixo x
plt.ylabel('y')                             #label eixo y  
plt.title('Função quadrática')              #label título
plt.legend()                                #mostrar a legenda dos dados
plt.show()
```

## Gráfico de dispersão

```python
import numpy as np
```

```python
x3 = np.linspace(0,10,30)
```

```python
print(x3)
```

```python
y3 = np.array(list(map(lambda x: x**2, x3)))
```

```python
print(y3)
```

```python
plt.scatter(x3,y3, color = 'blue', marker = 'o')
plt.xlabel('x')                             
plt.ylabel('y')                             
plt.title('Função quadrática') 
plt.legend(['$f(x) = x^2$'])
plt.show
```

```python
y3a = np.array(list(map(lambda x: x*10, x3)))
```

```python
plt.scatter(x3,y3, color = 'blue', marker = 'o')
plt.scatter(x3,y3a, color = 'red', marker = 'o')
plt.xlabel('x')                             
plt.ylabel('y')                             
plt.title('Função quadrática') 
plt.legend(['$f(x) = x^2$','$f(x) = 10x$'])
plt.grid()
plt.show
```

```python
x_scatter = np.arange(-10, 11, 1)  # Intervalo de -10 a 10 com passo 1
x_curve = np.linspace(-10, 10, 400)  # Intervalo de -10 a 10 com 400 pontos uniformemente espaçados

# Calculando f(x) = x^2 para ambos os conjuntos de x
y_scatter = x_scatter ** 2
y_curve = x_curve ** 2


plt.scatter(x_scatter, y_scatter, color='red')                     # Criando o gráfico scatter (pontos)

plt.plot(x_curve, y_curve, color='blue', label='$f(x) = x^2$')     # Criando a curva contínua


plt.title("Gráfico da Função f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
fig = plt.figure()         #criando a área gráfica.

#manipula posição dos gráficos através dos eixos [move-horiz,move-vert,fit-horiz,fit-vert]

ax1 = fig.add_axes([0.1,0.1,0.8,0.8])       #manipulando a área do gráfico 1
ax2 = fig.add_axes([0.2,0.5,0.4,0.3])       #manipulando a área do gráfico 1

ax1.plot(x3,y3,'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('quadrática')

ax2.plot(x3,y3a,'g')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('linear')
```

```python
# Diferentes estilos de Plots

# Dados


xxx = np.linspace(-0.75, 1., 100)
n = np.array([0,1,2,3,4,5])


# Subplots
fig, axes = plt.subplots(2, 2, figsize = (10,8))




# Plot 1
axes[0,0].scatter(xxx, xxx + 0.25 * np.random.randn(len(xxx)), color = "black")
axes[0,0].set_title("scatter")


# Plot 2
axes[0,1].step(n, n ** 2, lw = 2, color = "blue")
axes[0,1].set_title("step")


# Plot 3
axes[1,0].bar(n, n ** 2, align = "center", width = 0.5, alpha = 0.5, color = "magenta")
axes[1,0].set_title("bar")


# Plot 4
axes[1,1].fill_between(xx, xx ** 2, xx ** 3, alpha = 0.5, color = "green");
axes[1,1].set_title("fill_between");
```

```python
# Histogramas

# Dados
n = np.random.randint(0, 1000,(1,60)).flatten()

# Cria os subplots
fig, axes = plt.subplots(1, 2, figsize = (12,4))

# Plot 1
axes[0].hist(n)
axes[0].set_title("Histograma Padrão")
axes[0].set_xlim((min(n)-20, max(n)+20))

# Plot 2
axes[1].hist(n, cumulative = True, bins = 50)
axes[1].set_title("Histograma Cumulativo")
axes[1].set_xlim((min(n), max(n)))
```

## Gráficos em 3D

O Matplotlib oferece suporte para a criação de diversos tipos de representações gráficas em 3D, utilizando o módulo mpl_toolkits.mplot3d, que expande as capacidades da biblioteca para criar gráficos tridimensionais.

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()  # Cria uma nova figura

ax = fig.add_subplot(1,1,1, projection='3d')  # Adiciona um subplot 3D à figura

x = np.random.rand(100)  # Gera 100 valores entre 0 e 1 aleatórios para o eixo x
y = np.random.rand(100)  # Gera 100 valores entre 0 e 1 aleatórios para o eixo y
z = np.random.rand(100)  # Gera 100 valores entre 0 e 1 aleatórios para o eixo z

ax.scatter(x, y, z)  # Plota os pontos 3D
plt.show()  # Exibe o gráfico
```

```python
x = np.linspace(-5, 5, 100)  # Cria um vetor de 100 pontos de -5 a 5
y = np.linspace(-5, 5, 100)  # Cria um vetor de 100 pontos de -5 a 5
x, y = np.meshgrid(x, y)  # Cria uma grade 2D de coordenadas
z = np.sin(np.sqrt(x**2 + y**2))  # Calcula a função exponencial para cada ponto

fig = plt.figure()  # Cria uma nova figura
ax = fig.add_subplot(111, projection='3d')  # Adiciona um subplot 3D à figura
ax.plot_surface(x, y, z, cmap='viridis')  # Plota a superfície 3D
plt.show()  # Exibe o gráfico
```

```python
x = np.linspace(-5, 5, 100)  # Cria um vetor de 100 pontos de -5 a 5
y = np.linspace(-5, 5, 100)  # Cria um vetor de 100 pontos de -5 a 5
x, y = np.meshgrid(x, y)  # Cria uma grade 2D de coordenadas
z = np.sin(np.sqrt(x**2 + y**2))  # Calcula a função exponencial para cada ponto

fig = plt.figure()  # Cria uma nova figura
ax = fig.add_subplot(111, projection='3d')  # Adiciona um subplot 3D à figura
ax.contour3D(x, y, z, 50)  # Plota os contornos 3D com 50 níveis
plt.show()
```

```python
fig = plt.figure()  # Cria uma nova figura
ax = fig.add_subplot(111, projection='3d')  # Adiciona um subplot 3D à figura

_x = np.arange(4)  # Cria um vetor de 4 pontos para o eixo x
_y = np.arange(3)  # Cria um vetor de 3 pontos para o eixo y
_x, _y = np.meshgrid(_x, _y)  # Cria uma grade 2D de coordenadas
x, y = _x.ravel(), _y.ravel()  # Achata as matrizes em vetores 1D
z = np.zeros_like(x)  # Define a base das barras como zero
dx = dy = 0.5  # Define a largura das barras
dz = np.random.rand(12)  # Gera 12 alturas de barras aleatórias

ax.bar3d(x, y, z, dx, dy, dz, shade=True)  # Plota as barras 3D
plt.show()  # Exibe o gráfico
```

```python
fig = plt.figure()  # Cria uma nova figura
ax = fig.add_subplot(111, projection='3d')  # Adiciona um subplot 3D à figura

z = np.linspace(0, 15, 100)  # Cria um vetor de 100 pontos de 0 a 15
x = np.sin(z)  # Calcula a função seno para o eixo x
y = np.cos(z)  # Calcula a função cosseno para o eixo y

ax.plot(x, y, z)  # Plota a linha 3D
plt.show()
```
