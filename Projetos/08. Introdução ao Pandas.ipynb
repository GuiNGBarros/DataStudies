A função fillna() é usada para preencher os valores ausentes. A função oferece muitas opções. Podemos usar um valor específico, uma função agregada(por exemplo, a média) ou o valor anterior ou o seguinte.

```python
df3 = pd.read_csv("tabela_contabil_loja.csv")         #Importando um doc .csv para o pandas.
```

```python
df3.head(50)                         #Há ausência de dados (NaN) na coluna 'Quantidade'.
```

```python
df3.isna().sum()     #Fazendo a contagem dos NaNs de cada coluna.
```

```python
moda = df3['Quantidade'].value_counts()           #value_counts() mostra a quantidade repetida de cada valor.
```

```python
moda
```

```python
moda.index[0]         #A moda representa o valor com maior repetição.
```

```python
df3['Quantidade'].fillna(value = moda.index[0], inplace = True)       #fillna() preencheu os dados ausentes com o valor 
                                                                      # extraído da moda.
```

```python
df3.isna().sum()     #Fazendo novamente a contagem dos NaNs de cada coluna.
```

```python
df3.head(50)
```

```python
df3.describe()
```

Com o pandas criamos DataFrames, que são essencialmente tabelas. Como tal, podemos fazer consultas (query). 

```python
df4 = df3.copy()         #copiando o conteúdo do df3 no novo df4, para não correr o risco de perder a original.
```

```python
df4.query('800 < Valor_Venda < 4500')         #Consultando valores num intervalo do 'Valor_Venda'
```

```python
df4[df4.Nome_Produto == 'Guitarra']                #Consultando um produto em específico
```

```python
df4[(df4.Nome_Produto == 'Guitarra')&(df4.Quantidade == 5.0)]          #Consultando um produto E sua quantidade vendida.
```

```python
df4[(df4.Nome_Produto == 'Guitarra')|(df4.Nome_Produto == 'Violino')]        #Consultando um produto OU outro.
```

```python
df4[df4['Quantidade'].isin([1.0,3.0])]       #Filtrando os itens pela quantidade vendida. 'Quantidade' = 1. ou 3.
```

## Agrupamento de dados em DataFrames com GroupBy

A função pandas GroupBy é uma função versátil e fácil de usar que ajuda a obter uma visão geral dos dados. Isso torna mais fácil explorar o conjunto de dados e revelar o relacionamento entre as variáveis.

O código a seguir agrupará as linhas com base nas combinações Região/Cidade/Valor_Venda e nos dará a taxa média de vendas de cada grupo.

```python
df3[['Região','Cidade','Valor_Venda']].groupby(['Região','Cidade']).mean()
```

No exemplo abaixo uniremos a função groupby() com a função agg() para realizar agregação múltipla de métodos.

```python
df3[['Região','Cidade','Valor_Venda']].groupby(['Região','Cidade']).agg(['mean','count','sum'])
```
