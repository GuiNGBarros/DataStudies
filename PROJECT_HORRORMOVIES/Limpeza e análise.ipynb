```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.options.display.float_format = '{:.2f}'.format
```

```python
dfilmes = pd.read_csv("horrormovies.csv")
```

```python
dfilmes[['Gross', 'Budget']] = dfilmes[['Gross', 'Budget']].replace(0, np.nan)  
```

```python
dfilmes.isna().sum()
```

```python
dfilmes = dfilmes[~(dfilmes['Gross'].isna() | dfilmes['Budget'].isna())].reset_index(drop=True)
```

```python
dfilmes.isna().sum()
```

```python
display(dfilmes)
```

```python
# Função para tratar as barras verticais '|'
def tratar_barras(func):
    if pd.isna(func):
        return func  # Retorna o valor original se for NaN
    
    # Remove a barra vertical no final da string, se houver
    func = func.rstrip('|')
    
    # Substitui as barras verticais no meio por vírgulas
    func = func.replace('|', ', ')
    
    return func

# Aplica a função na coluna 'Directors'
dfilmes['Directors'] = dfilmes['Directors'].apply(tratar_barras)

# Aplica a função na coluna 'Writers'
dfilmes['Writers'] = dfilmes['Writers'].apply(tratar_barras)
```
