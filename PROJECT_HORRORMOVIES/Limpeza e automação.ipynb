## Automação para buscar os diretores dos filmes no site TMDB

```python
# Função para buscar o diretor de um filme no TMDB
def buscar_diretor(filme_title, api_key):
    # URL da requisição de busca de filmes no TMDB
    search_url = f'https://api.themoviedb.org/3/search/movie?api_key={api_key}&query={filme_title}'
    
    # Realiza a busca do filme por título
    response = requests.get(search_url)
    
    if response.status_code == 200:
        search_results = response.json()
        if search_results['results']:
            # Pega o ID do primeiro filme retornado
            movie_id = search_results['results'][0]['id']
            
            # URL da requisição para buscar detalhes do filme, incluindo o diretor
            details_url = f'https://api.themoviedb.org/3/movie/{movie_id}/credits?api_key={api_key}'
            response_details = requests.get(details_url)
            
            if response_details.status_code == 200:
                details = response_details.json()
                # Encontra o diretor entre os membros da equipe de produção (crew)
                for crew_member in details['crew']:
                    if crew_member['job'] == 'Director':
                        return crew_member['name']
    return None
```

```python
# Função para atualizar o dataframe 
def atualizar_diretores(dfilmes, linha_inicial, linha_final, api_key):
    # Verifica se a coluna 'director' já existe, senão, cria uma nova
    if 'director' not in dfilmes.columns:
        dfilmes['director'] = pd.NA

    # Itera sobre o intervalo de linhas especificado
    for i in range(linha_inicial, linha_final):
        filme_title = dfilmes.loc[i, 'title']  # Obtém o título do filme
        # Verifica se o diretor já foi preenchido, evita buscas repetidas
        if pd.isna(dfilmes.loc[i, 'director']):
            diretor = buscar_diretor(filme_title, api_key)  # Busca o diretor pelo título do filme
            dfilmes.at[i, 'director'] = diretor  # Atualiza o dataframe com o nome do diretor
    
    return dfilmes
```
